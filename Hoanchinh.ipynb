```python
import pandas as pd
import numpy as np


# Tạo bản sao và log
df_log = df_clean.copy()
df_log["median_income"] = np.log1p(df_log["median_income"])

print("Max trước log:", df_clean["median_income"].max())
print("Max sau log:", df_log["median_income"].max())
print("Skew trước log:", df_clean["median_income"].skew())
print("Skew sau log:", df_log["median_income"].skew())
```

1️⃣ Kết quả của bạn là gì?

Skew trước log: 1.01
Skew sau log: -3.10

Điều này nghĩa là:

Trước log: lệch phải nhẹ đến vừa (≈ 1)

Sau log: lệch trái rất mạnh (≈ -3)

👉 Tức là bạn đã “biến lệch phải thành lệch trái mạnh”.

2️⃣ Có phải ngoại lai không?

Skew ≠ Outlier.

Skew chỉ đo độ lệch phân phối.
Outlier được xác định bằng IQR hoặc Z-score.

Skew -3 không có nghĩa là có outlier nguy hiểm.
Nó chỉ nói rằng phân phối đang lệch trái mạnh.

3️⃣ Có cần xử lý tiếp không?

Với skew = -3:

Đây là lệch mạnh.

Log transform có thể đã “quá tay”.

Tuy nhiên:

Với mô hình hồi quy tuyến tính, không bắt buộc phải skew ≈ 0.

Điều quan trọng là giảm ảnh hưởng của giá trị cực lớn.

4️⃣ Tại sao lại ra skew âm lớn?

Thu nhập trung vị trong dataset này thực tế bị giới hạn trần (capped value).

Nhiều giá trị bằng mức tối đa → tạo “dồn cục” ở phía trên.

Sau log:

Phần đuôi phải bị nén mạnh.

Nhưng cụm giá trị trần làm phân phối lệch trái mạnh.

Đây là đặc điểm của dataset này, không phải lỗi.

5️⃣ Có nên xử lý tiếp không?

Trả lời ngắn gọn:

Không cần xử lý thêm nếu:

Bạn chỉ muốn giảm ảnh hưởng của giá trị lớn.

Bạn dùng mô hình hồi quy.

Chỉ cần ghi rõ trong báo cáo rằng:

“Sau log transformation, phân phối chuyển sang lệch trái do dữ liệu bị giới hạn trần ở mức cao, tuy nhiên ảnh hưởng của các giá trị cực lớn đã được giảm đáng kể.”

Khoảng giá trị thu hẹp đáng kể

Điều này cho thấy:

→ Log transform đã nén các giá trị lớn lại
→ Giảm độ phân tán
→ Giảm ảnh hưởng của outliers

Bạn có:

Nguyên nhân chính:

Dataset này có giá trị thu nhập bị "capped" ở mức 150001.

Sau khi log:

Các giá trị lớn bị dồn lại gần nhau

Phần dưới vẫn trải dài
→ tạo lệch trái

giúp quá trình học “cân bằng” hơn giữa các biến

Decision Tree không dựa vào khoảng cách hay độ lớn tuyệt đối của biến, mà dựa vào các ngưỡng chia.

Ví dụ cây có thể học kiểu:

nếu median_income < 3.5 thì rẽ trái

nếu total_rooms > 2500 thì rẽ phải

Cây chỉ quan tâm:

giá trị nào lớn hơn / nhỏ hơn ngưỡng nào

thứ tự tương đối của dữ liệu

Nếu bạn scale dữ liệu, ví dụ:

từ total_rooms > 2500

thành z-score > 0.42

thì bản chất chia nhánh vẫn y như vậy.
Tức là scaling không làm Decision Tree tốt lên đáng kể.

Random Forest là tập hợp nhiều Decision Tree.

ĐÁNH GIÁ HIỆU SUẤT MÔ HÌNH

```python
#Ví dụ tính các chỉ số đánh giá:
# RainForest Regressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
mae_rf = mean_absolute_error(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(y_test, y_pred_rf)

# Linear Regression
mae_lr = mean_absolute_error(y_test, y_pred_lr)
mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(mse_lr)
r2_lr = r2_score(y_test, y_pred_lr)

# Decision Tree
mae_dt = mean_absolute_error(y_test, y_pred_dt)
mse_dt = mean_squared_error(y_test, y_pred_dt)
rmse_dt = np.sqrt(mse_dt)
r2_dt = r2_score(y_test, y_pred_dt)
```
